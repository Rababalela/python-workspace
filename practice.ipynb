```python
def calculate_average(test, assignment, exam):
  final_mark = (test * 0.3) + (assignment * 0.2) + (exam * 0.5)
  return final_mark

def determine_result(final_mark):
  if final_mark >= 50:
    return "Pass"
  else:
     return "Fail"

# main program
test = float(input("Enter test mark:  "))
assignment = float(input("Enter assignment mark:  "))
exam = float(input("Enter exam mark:  "))

final = calculate_average(test, assignment, exam)
result = determine_result(final)

print(f"Final mark: {final}")
print(f"Result: {result}")
```

```python
def calculate_nett_salary(basic, bonus,deduction):
  nett_salery = (basic + bonus) - deduction
  return nett_salery

def determine_tax_bracket(nett_salery):
  if nett_salery < 5000:
    return "low bracket"

  elif nett_salery <= 15000:
    return "Middle bracket"

  else:
    return "High bracket"

# main program
basic = float(input("Enter basic salery:  "))
bonus = float(input("Enter bonus amount:  "))
deduction = float(input("Enter total deduction"))

income = calculate_nett_salary(basic, bonus, deduction)
tax = determine_tax_bracket(income)

print(f"the the nett salery {income:.2f}")
print(f"the tax bracket is {tax}")
```
